```python
#importing Libraies
from PIL import Image
import numpy as np
```

```python
import os
#training data
def training_data(filename,label):
    file1=os.listdir("C:\\Users\\arulkumar\\Desktop\\MLGuvi\\Guvi_final\\archive\\dataset\\training_set")
    file_path=os.path.join("C:\\Users\\arulkumar\\Desktop\\MLGuvi\\Guvi_final\\archive\\dataset\\training_set",filename)
    target=[]
    Means=[]
    for i in os.listdir(file_path) :
        path=os.path.join(file_path,i)
        image=Image.open(path) 
        mean=np.mean(image,axis=(0,1)) 
        Means.append(mean) 
        target.append(label)
    return Means,target
```

```python
#testing data
def testing_data(filename,label):
    file1=os.listdir("C:\\Users\\arulkumar\\Desktop\\MLGuvi\\Guvi_final\\archive\\dataset\\test_set")
    file_path=os.path.join("C:\\Users\\arulkumar\\Desktop\\MLGuvi\\Guvi_final\\archive\\dataset\\test_set",filename)
    target=[]
    Means=[]
    for i in os.listdir(file_path) :
        path=os.path.join(file_path,i)
        image=Image.open(path) 
        mean=np.mean(image,axis=(0,1)) 
        Means.append(mean) 
        target.append(label)
    return Means,target
```

```python
cat_train,cat_target=training_data("cats",0)
cat_target
```

```python
dog_train,dog_target=training_data("dogs",1)
dog_target
```

```python
cat_test,cat_target1=testing_data("cats",0)
cat_test
```

```python
dog_test,dog_target1=testing_data("dogs",1)
dog_test
```

```python
import pandas as pd
df=pd.DataFrame(cat_train)
df1=pd.DataFrame(dog_train)
df2 = pd.concat([df, df1], axis=0)
df2.tail()
```

```python
df3=pd.DataFrame(cat_target)
df4=pd.DataFrame(dog_target)
df3.tail()
```

```python
df5 = pd.concat([df3, df4], axis=0)
df5.tail()
```

```python
df_cat=pd.DataFrame(cat_test)
df_dog=pd.DataFrame(dog_test)
df_1 = pd.concat([df_cat, df_dog], axis=0)
df_1
```

```python
df_3=pd.DataFrame(cat_target1)
df_4=pd.DataFrame(dog_target1)
```

```python
df_5 = pd.concat([df_3, df_4], axis=0)
df_5
```

```python
df_test= pd.concat([df_1,df_5], axis=1,ignore_index=True)
df_test
```

```python
x_train=df2
y_train=df5
x_test=df_1
y_test=df_5
classifiers=[]
accuracy=[]
```

```python
#Decision Tree Classifier
### from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
plt.figure(figsize=(4,4),dpi=800)
plot_tree(classifier)
classifiers.append("Decision Tree classifier")
accuracy.append(result2)
```

```python
#Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
classifiers.append("Random Forest classifier")
accuracy.append(result2)
```

```python
#Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
gradient_booster = GradientBoostingClassifier(learning_rate=0.1)
gradient_booster.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
classifiers.append("Gradient Boosting Classifier")
accuracy.append(result2)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
classifiers.append("Logistic Regression")
accuracy.append(result2)
```

```python
#SVM
from sklearn.svm import SVC
svm = SVC(kernel="linear", gamma=0.5, C=1.0)
# Trained the model
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
classifiers.append("SVM")
accuracy.append(result2)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(30, 10))
plt.bar(classifiers,[i*100 for i in accuracy],width=0.5)
plt.title("Accuracy of each classifier")
plt.xlabel("Classifiers")
plt.ylabel("Accuracy")
plt.show()
```
